## 5.2 Euclidean Distance 

```python
from sklearn.metrics import pairwise_distances
```

```python
user_euclidean_pivot=pd.pivot_table(data=Book2,values='Book.Rating',index='User.ID',columns='Book.Title')
user_euclidean_pivot.index=Book2['User.ID'].unique()
user_euclidean_pivot
```

```python
user_euclidean_pivot2=user_euclidean_pivot.fillna(0)
user_euclidean_pivot2
```

```python
euclideon_distance_user = pairwise_distances(X=user_euclidean_pivot2, metric='euclidean')
euclideon_distance_user_df=pd.DataFrame(data=euclideon_distance_user)
euclideon_distance_user_df_index=Books['User.ID'].unique()
euclideon_distance_user_df_column=Books['User.ID'].unique()
euclideon_distance_user_df
```

## 5.3 Cosine Metrics

```python
cosine_metrics_user=1-pairwise_distances(X=user_euclidean_pivot2, metric='cosine')
cosine_metrics_user_df=pd.DataFrame(cosine_metrics_user)
cosine_metrics_user_df_index=Books['User.ID'].unique()
cosine_metrics_user_df_column=Books['User.ID'].unique()
cosine_metrics_user_df
```
